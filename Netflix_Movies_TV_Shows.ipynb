```python
#plot a bar graph Count vs Type
netflix.type.value_counts().plot(kind="bar",color="blue")
plt.show()
```

```python
#plot the number of movies each director has done.
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)
sns.countplot(data=netflix, x = netflix["director"], palette = 'mako',
             order = netflix['director'].value_counts().index[0:50])
plt.show()
```

```python
#plot the number of movies against the country.
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)
sns.countplot(data=netflix, x = netflix["country"], palette = 'rocket_r',
             order = netflix['country'].value_counts().index[0:50])
plt.show()
```

```python
#Check for the null values
netflix.date_added.isnull().sum()
```

```python
date = netflix[["date_added"]].dropna()
```

```python
#split the date_added into months and year
date['year'] = date["date_added"].apply(lambda x : x.split(',')[1])
date['month']= date["date_added"].apply(lambda x : x.lstrip().split(' ')[0])
```

```python
plt.xticks(rotation=45)
sns.countplot(x=date["year"],palette = "rocket")
plt.show()
```

```python
#Pie charts movie and tv show
labels = netflix["type"].value_counts().index
sizes = netflix["type"].value_counts()
plt.figure(figsize = (7,7))
plt.pie(sizes,labels=labels, rotatelabels = False, autopct = '%0.1f%%',textprops={'fontsize': 15})
plt.title('Type',color = 'purple', fontsize = 20)
plt.show()
```

```python
netflix_movie_only = netflix[netflix["type"]=="Movie"]
netflix_movie_only.head()
```

```python
#filtering short movies
time = netflix_movie_only[['duration']].dropna()
time['min'] = time["duration"].apply(lambda x : x.split(' ')[0])
time.head()
```

```python
#filtering the short movies
short_movies = time[time["min"] > '90']
short_movies.head()
```

```python
#plot the bar graph
plt.figure(figsize=(10,10))
plt.xticks(rotation = 90)
sns.countplot(x=short_movies["min"],palette = "rocket")
plt.show()
```
